```python
import os

original_file = 'original.txt'
altered_files_dir = 'altered'

altered_files = []
for file in os.listdir(altered_files_dir):
    if file.endswith('.txt'):
        altered_files.append(os.path.join(altered_files_dir, file))

with open(original_file, 'r') as f:
    original = f.readlines()



blocks = {}
for altered_file in altered_files:
    with open(altered_file, 'r') as f:
        altered = f.readlines()
    
    blocks[altered_file] = []
    diffs = []
    for i in range(len(original)):
        if original[i] != altered[i]:
            diffs.append(i)
    
    block = [diffs[0]]
    for i in range(1, len(diffs)):
        if diffs[i] - diffs[i-1] != 1:
            block.append(diffs[i-1])
            blocks[altered_file].append(block)
            block = [diffs[i]]
    block.append(diffs[-1])
    blocks[altered_file].append(block)
    
```

```python
blocks
```

```python
all_blocks = sorted([b for v in blocks.values() for b in v], key=lambda x: x[0])
for i in range(len(blocks)-1):
    if all_blocks[i][-1] >= all_blocks[i+1][0]:
        print('Overlapping blocks')
        break
else:
    print('No overlapping blocks')
```

```python
merged_text = original.copy()

for file, file_blocks in blocks.items():
    with open(file, 'r') as f:
        altered = f.readlines()
    for block in file_blocks:
        merged_text[block[0]:block[1]] = altered[block[0]:block[1]]    
```

```python
with open('merged.txt', 'w') as f:
    f.writelines(merged_text)
```

```python
from modeller import Environ, Selection, ModellerError
from modeller.automodel import LoopModel, refine

# Define a task ID directly for Jupyter Notebook (useful for parallel tasks if needed)
taskid = 1  # Set this to any number if you want to generate multiple models

# Initialize the MODELLER environment with a specific random seed for reproducibility
env = Environ(rand_seed=-1000-taskid)

class MyLoop(LoopModel):
    def select_loop_atoms(self):
        # Define the loop regions to refine
        rngs = (
            self.residue_range('93:X', '100:X'),
            self.residue_range('136:X', '145:X'),
            self.residue_range('171:X', '181:X'),
            self.residue_range('217:X', '223:X'),
            self.residue_range('254:X', '257:X'),
        )
        # Validate loop length (optional)
        for rng in rngs:
            if len(rng) > 30:
                raise ModellerError("Loop too long: loops must be ≤ 30 residues.")
        # Check if residues have topology
        s = Selection(rngs)
        if len(s.only_no_topology()) > 0:
            raise ModellerError("Some selected residues have no topology.")
        return s

# Input PDB file
input_pdb = 'modeller_7VV5_R-7T6T_R-7XJL_F-8WSS_R.pdb'

# Create the loop model
m = MyLoop(env, inimodel=input_pdb, sequence='loop')
m.set_output_model_format('PDB')  # Ensure output is in PDB format
m.loop.md_level = refine.slow     # Use a high refinement level for accuracy
m.loop.starting_model = m.loop.ending_model = taskid  # Set the task ID for parallelization

# Run the modeling process
m.make()

print(f"Loop refinement completed for task ID {taskid}.")
```
